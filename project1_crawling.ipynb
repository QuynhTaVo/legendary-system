```python
import requests, re, gzip
data_url = 'http://insideairbnb.com/get-the-data/'
```

```python
#crawling download links
content = requests.get(data_url)
content.encoding = 'utf-8'
download_links = re.findall('http:\/\/data\.insideairbnb\.com\/(?:(?!http:\/\/data\.insideairbnb\.com\/).)+?gz', content.text)
download_links
```
